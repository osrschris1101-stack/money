```python
## In this cell you reach X and then connect mannualy to the account, once done we run the next cell 

from seleniumbase import Driver
import time
import random
import pandas as pd
from selenium.webdriver.common.by import By
from transformers import pipeline
import yt_dlp
import os
from datetime import datetime


print("🚀 Launching Chrome 144 using SeleniumBase UC Mode...")
driver = Driver(uc=True, headless=False)
driver.get('https://www.gmail.com')
print("✅ Reached Gmail")
time.sleep(4)
driver.get('https://www.x.com')
print("✅ Reached X")
```

```python


# Using a global list so you don't lose data if you stop the script
COLLECTED_VIDEOS = [] 
SEEN_IDS = set()

def parse_reach(text_value):
    text_value = text_value.strip().upper().replace(',', '')
    if not text_value: return 0
    multiplier = 1000 if 'K' in text_value else 1000000 if 'M' in text_value else 1
    clean_val = text_value.replace('K', '').replace('M', '')
    try:
        return float(clean_val) * multiplier
    except:
        return 0

def run_video_automation(driver, target_count=340, min_reach=200000):
    print(f"🕵️ Target: {target_count} videos | Min Reach: {min_reach}")
    print("---------------------------------------------------------")

    while len(COLLECTED_VIDEOS) < target_count:
        tweets = driver.find_elements(By.CSS_SELECTOR, 'article[data-testid="tweet"]')
        
        for tweet in tweets:
            try:
                # 1. Video Check (Only proceed if it's a video)
                video_elements = tweet.find_elements(By.CSS_SELECTOR, '[data-testid="videoPlayer"], video')
                if not video_elements:
                    continue 

                # 2. URL & ID
                link_elem = tweet.find_element(By.XPATH, './/a[contains(@href, "/status/")]')
                url = link_elem.get_attribute('href')
                tweet_id = url.split('/')[-1]

                if tweet_id in SEEN_IDS:
                    continue

                # 3. Reach Check
                lines = tweet.text.split('\n')
                reach_raw = lines[-1]
                reach_num = parse_reach(reach_raw)

                if reach_num >= min_reach:
                    # 4. Description
                    try:
                        desc = tweet.find_element(By.XPATH, './/div[@data-testid="tweetText"]').text
                    except:
                        desc = "No text content"

                    # 5. Save and Print Progress
                    COLLECTED_VIDEOS.append({
                        'url': url,
                        'reach': reach_num,
                        'description': desc,
                        'timestamp': time.strftime("%H:%M:%S")
                    })
                    SEEN_IDS.add(tweet_id)
                    
                    # --- PROGRESS PRINTING ---
                    current_len = len(COLLECTED_VIDEOS)
                    remaining = target_count - current_len
                    percent = (current_len / target_count) * 100
                    
                    print(f"🎬 NEW VIDEO ADDED!")
                    print(f"🔗 URL: {url}")
                    print(f"📊 Reach: {reach_raw}")
                    print(f"📈 Progress: {current_len}/{target_count} ({percent:.1f}%) | ⏳ Remaining: {remaining}")
                    print("-" * 30)

                if len(COLLECTED_VIDEOS) >= target_count:
                    break
                    
            except:
                continue

        # Stealth Scrolling
        scroll_dist = random.randint(1800, 3000)
        wait_time = random.uniform(4.5, 8.5)
        
        print(f"😴 Scrolling... (Searching for {target_count - len(COLLECTED_VIDEOS)} more)")
        driver.execute_script(f"window.scrollBy(0, {scroll_dist});")
        time.sleep(wait_time)

    return pd.DataFrame(COLLECTED_VIDEOS)

# --- EXECUTION ---
df_final = run_video_automation(driver, target_count=20, min_reach=80000)
df_final.to_csv('data_links_twitter.csv', index=False)
```
